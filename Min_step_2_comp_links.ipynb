```python
import pandas as pd
import string as st
```

```python
#Load file
links = pd.read_csv('2_component_links_compiled.csv')
#Making sure data type is a string
links.NEWSUD_format = links.NEWSUD_format.astype(str)
#Create new column for step number
links['step_number'] = 0
#Remove '\n' character that was on the end of some newsuds
links.NEWSUD_format = links.NEWSUD_format.apply(lambda x: x.replace("\n", " "))

```

```python
#Function to find the length of the 2-component link, ignoring lower-cases
def string_count (n): return len(filter(lambda x: x in st.uppercase, n))
```

```python
# Apply the previous function to the link table 
links.step_number = links.NEWSUD_format.apply(lambda y: string_count(y))
```
